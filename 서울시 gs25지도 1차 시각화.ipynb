```python
import urllib.request as req
from selenium import webdriver
from selenium.common import exceptions
from bs4 import BeautifulSoup
import folium
import time

driver = webdriver.Chrome('C:/Users/seob6/Desktop/crawling/chromedriver/chromedriver.exe')
url = 'https://store.naver.com/attractions/list?page={}&query=%EC%84%9C%EC%9A%B8%20GS25'
output = []
output2 = []
for i in range(1, 10):
    link = url.format(i)
    driver.get(link)
    driver.implicitly_wait(2)
    time.sleep(1.5)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    title = soup.select("div.tit a.name > span")
    address = soup.find_all("div", {"class":"txt ellp"})
   
    for t in title:
        output.append(t.get_text().strip())
    for a in address:
        output2.append(a.get_text().strip())
        
print(output)    
print(output2)
```

## googlemaps

```python
import googlemaps
import numpy as np
from tqdm import tqdm_notebook

lat = []
lng = []
for n in tqdm_notebook(gs.index):
    try:
        tmp_add = str(gs['주소'][n]).split('(')[0]
        tmp_map = gmaps.geocode(tmp_add)
        
        tmp_loc = tmp_map[0].get('geometry')
        lat.append(tmp_loc['location']['lat'])
        lng.append(tmp_loc['location']['lng'])
        
    except:
        lat.append(np.nan)
        lng.append(np.nan)
        print('Here is nan')

gs['lat'] = lat
gs['lng'] = lng
gs.head()    
```

```python
gs.to_csv("C:/Users/seob6/Desktop/DT/maru/GS_complete.csv")
```
